```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('/Users/damaso/Documents/Proyectos/java/IA/data/premierleague/epl_final.csv')
```

```python
def drop_labels(df: pd.DataFrame, class_name):
    return df.drop(class_name, axis=1)
```

```python
df.info()
```

```python
df.describe()
```

```python
df = drop_labels(df, ['Season', 'FullTimeHomeGoals', 'FullTimeAwayGoals', 'HalfTimeHomeGoals','HalfTimeAwayGoals','HalfTimeResult','HomeShots','AwayShots', 'HomeShotsOnTarget', 'AwayShotsOnTarget', 'HomeCorners',  'AwayCorners', 'HomeFouls','AwayFouls','HomeYellowCards','AwayYellowCards','HomeRedCards','AwayRedCards'])
df.head(5)
```

```python
df[['MatchDateYear', 'MatchDateMonth', 'MatchDateDay']] = df['MatchDate'].str.split('-', expand=True)
df = drop_labels(df, ['MatchDate'])
df.head(5)
```

```python
df = df[['MatchDateYear', 'MatchDateMonth', 'MatchDateDay', 'HomeTeam', 'AwayTeam','FullTimeResult']]
df.head(5)
```

```python
df.loc[df['FullTimeResult'] == 'H', 'FullTimeResult'] = 0
df.loc[df['FullTimeResult'] == 'D', 'FullTimeResult'] = 1
df.loc[df['FullTimeResult'] == 'A', 'FullTimeResult'] = 2
df
```

```python
df.to_csv('premier_league_data.csv', index=False)
```
